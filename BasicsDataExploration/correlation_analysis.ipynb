# Zadanie 1. Analiza wspolczynnikow korelacji

```python
!pip install scipy
```

```python
import pandas as pd
from sklearn.datasets import load_wine
import numpy as np
import scipy.stats as stats
# Importujemy jarque_berra jawnie z scipy.stats (Ten wiersz nie jest konieczny, jeśli dostęp jest przez stats)
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Wczytanie zestawu danych Wine i utworzenie DataFrame
print("Wczytywanie zestawu danych Wine...")
wine = load_wine()
df_wine = pd.DataFrame(data=wine.data, columns=wine.feature_names)
df_wine['target'] = wine.target
print("Zestaw danych Wine wczytany.")
print(df_wine.head())
print("\n")

# 2. Obliczanie i wyświetlanie podstawowych miar statystycznych
print("Obliczanie podstawowych miar statystycznych...")
statistical_summary = df_wine.describe()
print("Podstawowe miary statystyczne:")
display(statistical_summary)

# Dodatkowe miary statystyczne: mediana, skośność
median = df_wine.median()
skewness = df_wine.skew()

print("\nMediana:")
print(median)

print("\nSkośność:")
print(skewness)

# Zapisywanie miar statystycznych do pliku CSV
# Można wybrać txt lub xls, ale CSV jest najczęściej używany i łatwy do pracy w Colab
statistical_summary.to_csv('wine_statistical_summary.csv')
median.to_csv('wine_median.csv', header=['median'])
skewness.to_csv('wine_skewness.csv', header=['skewness'])
print("\nMiary statystyczne zapisano do plików CSV.")
print("\n")


# 3. Badanie normalności cech i stosowanie odpowiedniej miary korelacji

# Funkcja do sprawdzania normalności (Shapiro-Wilk i Normaltest jako alternatywa dla Jarque-Berra)
def check_normality(data, alpha=0.05):
    results = {}
    for col in data.columns:
        if data[col].dtype in ['int64', 'float64']: # Sprawdzamy tylko kolumny numeryczne
            # Dostęp do shapiro przez alias stats
            shapiro_test = stats.shapiro(data[col])
            # Używamy normaltest jako alternatywy dla Jarque-Berra
            normal_test = stats.normaltest(data[col])
            results[col] = {
                'Shapiro-Wilk_statistic': shapiro_test.statistic,
                'Shapiro-Wilk_p-value': shapiro_test.pvalue,
                'Normaltest_statistic': normal_test.statistic, # Zmieniono z Jarque-Berra
                'Normaltest_p-value': normal_test.pvalue,      # Zmieniono z Jarque-Berra
                'Is_Normal_(Shapiro_p>alpha)': shapiro_test.pvalue > alpha,
                'Is_Normal_(Normaltest_p>alpha)': normal_test.pvalue > alpha # Zmieniono z Jarque-Berra
            }
    return pd.DataFrame(results).T

print("Sprawdzanie normalności cech...")
normality_results = check_normality(df_wine.drop(columns=['target'])) # Wykluczamy 'target'
print("Wyniki testów normalności:")
display(normality_results)

# Określenie, którą miarę korelacji użyć (Pearson czy Spearman)
# Jeśli cecha nie jest rozkładem normalnym według obu testów, używamy Spearmana.
# W przeciwnym razie, jeśli co najmniej jeden test pokazuje normalność, używamy Pearsona.
# Dla uproszczenia i zapewnienia ogólnej porównywalności można również użyć Spearmana
# dla wszystkich par, jeśli istnieje co najmniej jedna zmienna o rozkładzie nienormalnym,
# lub obliczyć oba współczynniki i wyjaśnić wybór.
# Dla tego zadania, możemy określić, dla których par użyć Pearsona, a dla których Spearmana.

# Określamy, dla których cech użyć korelacji Pearsona
# Zaktualizowano nazwy kolumn w filtrowaniu na podstawie zmiany na Normaltest
pearson_cols = normality_results[(normality_results['Is_Normal_(Shapiro_p>alpha)'] == True) |
                                 (normality_results['Is_Normal_(Normaltest_p>alpha)'] == True)].index.tolist()

# Określamy, dla których cech użyć korelacji Spearmana (dla pozostałych)
spearman_cols = normality_results.index.difference(pearson_cols).tolist()

print(f"\nCechy, dla których można użyć korelacji Pearsona: {pearson_cols}")
print(f"Cechy, dla których prawdopodobnie lepiej użyć korelacji Spearmana: {spearman_cols}")

# Obliczanie macierzy korelacji
print("\nObliczanie macierzy korelacji...")

# Macierz korelacji Pearsona (dla par, gdzie obie cechy mogą mieć rozkład normalny)
# Jest to trudniejsze do automatycznego zaimplementowania dla każdej pary.
# Najprostsze podejście do demonstracji: obliczyć obie macierze.

# Macierz korelacji Pearsona
correlation_matrix_pearson = df_wine.drop(columns=['target']).corr(method='pearson')
print("\nMacierz korelacji Pearsona:")
display(correlation_matrix_pearson)

# Macierz korelacji Spearmana
correlation_matrix_spearman = df_wine.drop(columns=['target']).corr(method='spearman')
print("\nMacierz korelacji Spearmana:")
display(correlation_matrix_spearman)

# Wizualizacja macierzy korelacji za pomocą map ciepła
print("\nWizualizacja macierzy korelacji...")

# Mapa ciepła korelacji Pearsona
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix_pearson, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Mapa ciepła korelacji Pearsona dla zestawu danych Wine') # Podpis dla mapy ciepła Pearsona
plt.show()

# Mapa ciepła korelacji Spearmana
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix_spearman, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Mapa ciepła korelacji Spearmana dla zestawu danych Wine') # Podpis dla mapy ciepła Spearmana
plt.show()

print("\nAnaliza korelacji zakończona.")
```

# Zadanie 2: Analiza reszt za pomoca funkcji autokorelacji

```python
from sklearn.linear_model import LinearRegression
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.stats.stattools import durbin_watson
from statsmodels.stats.diagnostic import acorr_ljungbox

# 1. Wybierz dwie silnie ze sobą korelujące zmienne
# Na podstawie macierzy korelacji z Zadania 1, znajdujemy pary o wysokiej wartości bezwzględnej korelacji.
# Możemy przejrzeć 'correlation_matrix_pearson' lub 'correlation_matrix_spearman'.
# Sprawdźmy macierz Pearsona dla przykładu. Szukamy wartości bliskich 1 lub -1 (poza przekątną).

print("\nSzukanie silnie korelujących zmiennych...")
# Znajdźmy pary cech o największej wartości bezwzględnej korelacji (poza przekątną)
# Utwórzmy 'unstacked' serię z macierzy korelacji Pearsona
stacked_corr = correlation_matrix_pearson.unstack()
# Usuń pary cech z samymi sobą (korelacja = 1)
sorted_corr = stacked_corr.sort_values(ascending=False)
# Usuń duplikaty par (a z b jest to samo co b z a)
unique_pairs = sorted_corr[sorted_corr.index.get_level_values(0) < sorted_corr.index.get_level_values(1)]

# Weźmy najbardziej pozytywnie lub negatywnie skorelowaną parę
strongest_pair = unique_pairs.abs().idxmax()
strongest_corr_value = unique_pairs[strongest_pair]

var1_name, var2_name = strongest_pair
print(f"Wybrano silnie korelujące zmienne: '{var1_name}' i '{var2_name}' (Korelacja Pearsona: {strongest_corr_value:.2f})")

# Przygotowanie danych dla regresji
# Zmienna niezależna (predictor)
X = df_wine[[var1_name]]
# Zmienna zależna (response)
y = df_wine[var2_name]

# 2. Sporządź wykres rozrzutu ilustrujący związek między zmiennymi.
plt.figure(figsize=(8, 6))
sns.scatterplot(x=var1_name, y=var2_name, data=df_wine)
plt.title(f'Wykres rozrzutu między "{var1_name}" a "{var2_name}"')
plt.xlabel(var1_name)
plt.ylabel(var2_name)
plt.grid(True)
plt.show()

# 3. Budowa Modelu Regresji Liniowej
print("\nBudowa modelu regresji liniowej...")
model = LinearRegression()
model.fit(X, y)

print(f"Model dopasowany: y = {model.coef_[0]:.4f} * {var1_name} + {model.intercept_:.4f}")

# Obliczenie przewidywanych wartości
y_pred = model.predict(X)

# Obliczenie reszt
residuals = y - y_pred
print("Reszty modelu obliczone.")

# 4. Analiza Reszt:

# a) Stwórz histogram reszt - wykres gęstości, aby ocenić rozkład reszt.
print("\nAnaliza reszt...")
plt.figure(figsize=(8, 6))
sns.histplot(residuals, kde=True)
plt.title('Histogram reszt z wykresem gęstości')
plt.xlabel('Wartość reszty')
plt.ylabel('Częstość')
plt.show()

# b) Oblicz i zwizualizuj funkcje autokorelacji (ACF) dla reszt do opóźnienia 50.
print("Wizualizacja funkcji autokorelacji (ACF) reszt...")
plt.figure(figsize=(10, 6))
# plot_acf domyślnie rysuje dla wszystkich dostępnych opóźnień.
# nlags=50 ograniczy to do opóźnienia 50.
plot_acf(residuals, lags=50, title='Funkcja autokorelacji (ACF) dla reszt')
plt.xlabel('Opóźnienie (Lag)')
plt.ylabel('Współczynnik autokorelacji')
plt.grid(True)
plt.show()
print("Wykres ACF wygenerowany.")

# c) Wykonaj test Durbin-Watson: Sprawdź obecność autokorelacji w resztach.
print("\nWykonanie testu Durbin-Watson...")
dw_statistic = durbin_watson(residuals)
print(f"Statystyka Durbin-Watsona: {dw_statistic:.4f}")
# Interpretacja: Wartości bliskie 2 sugerują brak autokorelacji.
# Wartości znacznie mniejsze od 2 sugerują pozytywną autokorelację.
# Wartości znacznie większe od 2 sugerują negatywną autokorelację.
# Wartości bliskie 0 lub 4 sugerują silną autokorelację.

# d) Wykonaj test Ljung-Boxa: Przeprowadź test na obecność autokorelacji dla kilku opóźnień.
print("\nWykonanie testu Ljung-Boxa dla kilku opóźnień...")
# Przeprowadzamy test dla kilku opóźnień, np. [10, 20, 30]
lags_to_test = [10, 20, 30] # Możesz dostosować te opóźnienia
ljung_box_results = acorr_ljungbox(residuals, lags=lags_to_test, return_df=True)
print("Wyniki testu Ljung-Boxa:")
print(ljung_box_results)
# Interpretacja: Jeśli p-wartość (p-value) dla danego opóźnienia jest mniejsza od wybranego poziomu istotności (np. 0.05),
# odrzucamy hipotezę zerową o braku autokorelacji dla tego opóźnienia, co sugeruje obecność autokorelacji.

print("\nAnaliza reszt zakończona.")
```

# Zadanie 3: Wyrownanie sygnalow za pomoca korelacji krzyzowej

```python
import scipy.io
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns # do wizualizacji

# 1. Wczytaj dane 'relatedsig.mat'
# Plik relatedsig.mat powinien być dostępny w tym samym katalogu co notebook, lub musisz podać pełną ścieżkę
plik_mat = '/content/drive/MyDrive/relatedsig.mat' # Używamy ścieżki z wyjścia użytkownika
data = None # Initialize data to None

try:
    data = scipy.io.loadmat(plik_mat)
    print(f"Plik '{plik_mat}' wczytany pomyślnie.")
    # Sprawdźmy klucze wczytanych danych
    print("Klucze w wczytanych danych:", data.keys())
except FileNotFoundError:
    print(f"Błąd: Plik '{plik_mat}' nie znaleziono. Upewnij się, że plik znajduje się w prawidłowej lokalizacji.")
    # Do not exit, allow the script to continue but handle the missing data

# Check if data was successfully loaded
if data is not None:
    # Try to load signals using the correct keys found in the traceback
    # Based on the traceback, the keys are 's1', 's2', 's3' or 'sig1', 'sig2'
    # Let's assume 's1', 's2', 's3' are the desired signals for this example
    signal1 = None # Initialize signals to None
    signal2 = None
    signal3 = None

    try:
        # Use the correct keys 's1', 's2', 's3' based on the traceback output
        signal1 = data.get('s1', np.array([])).flatten() # Use .get with a default empty array
        signal2 = data.get('s2', np.array([])).flatten()
        signal3 = data.get('s3', np.array([])).flatten()
        print("Sygnały wczytane z danych.")
        print(f"Długość sygnału 1 ('s1'): {len(signal1)}")
        print(f"Długość sygnału 2 ('s2'): {len(signal2)}")
        print(f"Długość sygnału 3 ('s3'): {len(signal3)}")

        # Check if signals are empty before proceeding with calculations and plotting
        if len(signal1) > 0 and len(signal2) > 0 and len(signal3) > 0:
            # 2. Oblicz korelacje krzyżowe pomiędzy trzema parami sygnałów.
            # Znormalizuj je tak, aby ich maksymalna wartość wynosiła jeden.
            # Polożenie maksymalnych wartości korelacji krzyżowych wskazuje na przewagę lub opóźnienie czasowe.

            print("\nObliczanie korelacji krzyżowych...")

            # Funkcja pomocnicza do obliczania i normalizacji korelacji krzyżowej
            def calculate_and_normalize_cross_correlation(signal_a, signal_b):
                # Oblicz korelacje krzyżową
                # 'full' daje wszystkie możliwe przesunięcia
                cross_corr = np.correlate(signal_a, signal_b, mode='full')

                # Znormalizuj, dzieląc przez maksymalną wartość bezwzględną
                max_corr_abs = np.max(np.abs(cross_corr))
                if max_corr_abs > 0:
                    normalized_cross_corr = cross_corr / max_corr_abs
                else:
                    normalized_cross_corr = cross_corr # Unikaj dzielenia przez zero, jeśli korelacja wynosi 0

                # Oblicz opóźnienia (lags)
                # Długość korelacji krzyżowej dla mode='full' wynosi len(a) + len(b) - 1
                n_a = len(signal_a)
                n_b = len(signal_b)
                lags = np.arange(-(n_b - 1), n_a) # Opóźnienia od -(n_b-1) do n_a-1

                return lags, normalized_cross_corr

            # Oblicz korelacje dla par: (1,2), (1,3), (2,3)
            lags_1_2, corr_1_2 = calculate_and_normalize_cross_correlation(signal1, signal2)
            lags_1_3, corr_1_3 = calculate_and_normalize_cross_correlation(signal1, signal3)
            lags_2_3, corr_2_3 = calculate_and_normalize_cross_correlation(signal2, signal3)

            print("Korelacje krzyżowe obliczone i znormalizowane.")

            # Znajdź opóźnienie (lag) odpowiadające maksymalnej korelacji dla każdej pary
            # Ensure the correlation arrays are not empty before finding argmax
            if len(corr_1_2) > 0:
                lag_1_2 = lags_1_2[np.argmax(np.abs(corr_1_2))]
                print(f"\nOpóźnienie dla pary (Sygnał 1, Sygnał 2): {lag_1_2}")
            else:
                 print("\nKorelacja dla pary (Sygnał 1, Sygnał 2) jest pusta.")
                 lag_1_2 = None # Or some other indicator

            if len(corr_1_3) > 0:
                lag_1_3 = lags_1_3[np.argmax(np.abs(corr_1_3))]
                print(f"Opóźnienie dla pary (Sygnał 1, Sygnał 3): {lag_1_3}")
            else:
                 print("Korelacja dla pary (Sygnał 1, Sygnał 3) jest pusta.")
                 lag_1_3 = None

            if len(corr_2_3) > 0:
                lag_2_3 = lags_2_3[np.argmax(np.abs(corr_2_3))]
                print(f"Opóźnienie dla pary (Sygnał 2, Sygnał 3): {lag_2_3}")
            else:
                 print("Korelacja dla pary (Sygnał 2, Sygnał 3) jest pusta.")
                 lag_2_3 = None


            # Interpretacja opóźnień:
            # Dodatnie opóźnienie 'lag_a_b' oznacza, że sygnał B jest przesunięty w prawo (opóźniony)
            # względem sygnału A o 'lag_a_b' próbek.
            # Ujemne opóźnienie 'lag_a_b' oznacza, że sygnał B jest przesunięty w lewo (wyprzedza)
            # względem sygnału A o |lag_a_b| próbek.

            # 3. Wykreśl korelacje krzyżowe. Na każdym wykresie pokaż położenie maksimum.

            print("\nWykresy korelacji krzyżowych...")

            plt.figure(figsize=(15, 10))

            # Wykres korelacji (Sygnał 1, Sygnał 2)
            plt.subplot(3, 1, 1) # 3 wiersze, 1 kolumna, 1. wykres
            plt.plot(lags_1_2, corr_1_2)
            if lag_1_2 is not None: # Only plot if lag was found
                plt.axvline(lag_1_2, color='r', linestyle='--', label=f'Maksimum przy opóźnieniu {lag_1_2}')
            plt.title('Korelacja krzyżowa: Sygnał 1 i Sygnał 2')
            plt.xlabel('Opóźnienie (próbki)')
            plt.ylabel('Znormalizowana korelacja')
            plt.grid(True)
            plt.legend()

            # Wykres korelacji (Sygnał 1, Sygnał 3)
            plt.subplot(3, 1, 2) # 3 wiersze, 1 kolumna, 2. wykres
            plt.plot(lags_1_3, corr_1_3, color='g')
            if lag_1_3 is not None: # Only plot if lag was found
                plt.axvline(lag_1_3, color='r', linestyle='--', label=f'Maksimum przy opóźnieniu {lag_1_3}')
            plt.title('Korelacja krzyżowa: Sygnał 1 i Sygnał 3')
            plt.xlabel('Opóźnienie (próbki)')
            plt.ylabel('Znormalizowana korelacja')
            plt.grid(True)
            plt.legend()

            # Wykres korelacji (Sygnał 2, Sygnał 3)
            plt.subplot(3, 1, 3) # 3 wiersze, 1 kolumna, 3. wykres
            plt.plot(lags_2_3, corr_2_3, color='purple')
            if lag_2_3 is not None: # Only plot if lag was found
                plt.axvline(lag_2_3, color='r', linestyle='--', label=f'Maksimum przy opóźnieniu {lag_2_3}')
            plt.title('Korelacja krzyżowa: Sygnał 2 i Sygnał 3')
            plt.xlabel('Opóźnienie (próbki)')
            plt.ylabel('Znormalizowana korelacja')
            plt.grid(True)
            plt.legend()

            plt.tight_layout() # Poprawia rozmieszczenie wykresów, aby się nie nakładały
            plt.show()

            # 4. Zaznacz odpowiednio opóźnienia.
            # Zaznaczenie opóźnień zostało zrealizowane na wykresach za pomocą pionowych linii (plt.axvline)
            # oraz w legendzie wykresu i w wydruku konsoli.

            print("\nAnaliza korelacji krzyżowej zakończona.")

            # Dodatkowe informacje do omówienia podczas obrony:
            # - Jak interpretować wykresy korelacji krzyżowej (kształt, wysokość maksimum).
            # - Znaczenie położenia maksimum (opóźnienie czasowe).
            # - Jak wykorzystać te opóźnienia do wyrównania sygnałów (np. przesunięcie sygnałów).
            # - Dlaczego korelacja krzyżowa jest przydatna do synchronizacji sygnałów asynchronicznych.
        else:
            print("\nNie można kontynuować analizy korelacji krzyżowej, ponieważ co najmniej jeden sygnał jest pusty (klucze 's1', 's2', 's3' mogły nie zawierać danych).")


    except KeyError as e:
        # This block is now redundant because we are using .get(), but leaving it
        # as a fallback or for clarity that we expected specific keys.
        print(f"Błąd: Klucz {e} nie znaleziono w pliku '{plik_mat}'. Sprawdź, jakie zmienne są zapisane w pliku .mat.")
        print("Dostępne klucze:", data.keys())
        print("\nNie można kontynuować analizy korelacji krzyżowej, ponieważ nie znaleziono oczekiwanych sygnałów ('s1', 's2', 's3').")


else:
    print("\nNie można kontynuować analizy korelacji krzyżowej, ponieważ plik danych nie został wczytany.")
```
